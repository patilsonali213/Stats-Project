```python
# Ques.2 Statistical Analysis:

from scipy.stats import ttest_ind, chi2_contingency
# In Statistics, t-test is a test which is used for hypothesis testing. In this we take sample from each sets and assume a null hypothesis in which 2 means are equal and according to formula values are generated.

# Example: T-test for sales in different regions
Ho= "Sales are same in different region"
Ha= "Sales are not samew in different region"

x= np.array(df[df.Sales=="yes"].Region)
y= np.array(df[df.Sales=="no"].Region)

t, p_value=ttest_ind(x,y,axis=0)
if p_value<0.5:
    print(f'{Ha} as the p_value ({p_value})<0.5')
else:
    print(f'{Ho} as the p_value ({p_value})>0.5')
          

 
```

```python
# Example: Chi-square test for independence between Ship Mode and Customer Segment

# chi-square test for independence shows that whether the variables are independednt or not.

Ho= "Ship mode not independent to Customer Segment"
Ha= "Ship mode is independent to Customer Segment"

crosstab= pd.crosstab(df['Ship Mode'], df['Customer Segment'])
chi, p_value, dof, expected= chi2_contingency(crosstab)
if p_value<0.5:
    print(f'{Ha} as the p_value ({p_value.round(3)})<0.5')
else:
    print(f'{Ho} as the p_value ({p_value.round(3)})>0.5')
```
